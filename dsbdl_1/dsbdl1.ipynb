```python
import numpy as np
import pandas as pd 
```

```python
dataset = pd.read_csv("Data.csv")
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
dataset = pd.get_dummies(dataset,columns=["Country"],prefix='',prefix_sep='',dtype=int)
dataset = pd.get_dummies(dataset,columns=["Purchased"],prefix='',prefix_sep='',dtype=int)
dataset
```

```python
col_means = dataset.mean()
dataset =dataset.replace(to_replace=np.nan,value=col_means)
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dataset.iloc[:,:2] = scaler.fit_transform(dataset.iloc[:,:2])
dataset
```
